```python
#1. Line Graph/plot using LISTS
import matplotlib.pyplot as plt
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]

#plt.plot(x-axis, y-axis)
plt.plot(a, b, color = 'lime', marker= '*')
plt.title('LINE PLOT')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
#Line graph using numpy array
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])
plt.plot(x, y, marker = '*', color = 'lime')
plt.title('LINE PLOT')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks(range(0, 30, 2))
```

```python
#2. Scatter plot
a = [1, 2, 3, 4, 5]
b = [6, 7, 8, 9, 10]
plt.scatter(a, b, color = ['gold', 'red', 'lime', 'blue', 'green'])
plt.plot(a, b)
plt.title('Scatter plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
#3. BAR GRAPH
names = ['Akshay', 'Dheeraj', 'John', 'Mac']
weight = [58, 86, 78, 69]
plt.barh(names, weight, color = ['Red', 'Orange', 'Lime', 'gold'])
#change  plt.bar to bar.h for horizontal bars
plt.title('Bar Graph')
plt.xlabel('Names')
plt.ylabel('Weight')
```
